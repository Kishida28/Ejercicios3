```python
import pandas as pd
```

```python
#sheet_name verifique el nombre de la pestaña en la hoja de calculo Minedu-Matricula-2019.xlsx
df_tendencia = pd.read_excel('archivo.xlsx',
                       sheet_name='Hoja04',
                       skiprows=0,
                       skipfooter=0)

print ('Datos leídos en un marco de datos de pandas!')
```

```python
df_tendencia.head(5)
```

```python
# imprime el tamaño del dataframe.
print(df_tendencia.shape)
```

```python
# examinemos los tipos de etiquetas de columna
all(isinstance(column, str) for column in df_tendencia.columns)
```

```python
df_tendencia.columns = list(map(str, df_tendencia.columns))

# vamos a ver los tipos de etiquetas de columna ahora
all(isinstance(column, str) for column in df_tendencia.columns)
```

```python
df_tendencia.set_index('REGION', inplace=True)

# veamos los primeros cinco elementos y veamos cómo se cambió el marco de datos (dataframe) Country (país) resaltado
df_tendencia.head()
```

```python
# imprime el tamaño del dataframe.
print(df_tendencia.shape)
```

```python
# La suma de las filas: df_tendencia.sum(axis=1)
# df_tendencia.sum(axis=1) / (len(df_tendencia.columns)-1)   ## -1 para restar la columna Region

df_tendencia['Promedio'] = (len(df_tendencia.columns)-1)

# veamos los primeros cinco elementos y comprobemos cómo se cambió el marco de datos (dataframe)
df_tendencia.head()
```

```python
print ('Dimensiones de los datos:', df_tendencia.shape)
```

```python
# finalmente, creemos una lista de años desde 2001 hasta 2015
# esto será útil cuando comencemos a graficar los datos
años = list(map(str, range(2001, 2016)))  # note habilitar a un año adicional

años
```

```python
# utilice inline al final de la línea para generar los gráficos dentro del navegador
%matplotlib inline 

import matplotlib as mpl
import matplotlib.pyplot as plt

mpl.style.use('ggplot') # opcional: para mostrar el estilo ggplot-like 

# chequeando la última versión de Matplotlib
print ('Matplotlib version: ', mpl.__version__) # >= 2.0.0
```

```python
df_tendencia.sort_values(['REGION'], ascending=True, axis=0, inplace=True)

# obteniendo las 5 primeros años
df_top5 = df_tendencia.head(5)    #pudo tambien ser tail(5) para seleccionar los 5 últimos

# transpose the dataframe
df_top5 = df_top5[años].transpose() 

df_top5.head()
```

```python
df_top5.index = df_top5.index.map(int) # let's change the index values of df_top5 to type integer for plotting
df_top5.plot(kind='area', 
             stacked=False,
             figsize=(20, 10), # pass a tuple (x, y) size
             )

plt.title('Tendencia de conclusión, primaria, grupo de edades 12-13 (% del total)')
plt.ylabel('Porcentaje')
plt.xlabel('Años')

plt.show()
```

```python
df_top5.index = df_top5.index.map(int) # let's change the index values of df_top5 to type integer for plotting
df_top5.plot(kind='area', 
             alpha=0.25, # 0-1, default value a= 0.5
             stacked=False,
             figsize=(20, 10), # pass a tuple (x, y) size
             )

plt.title('Tendencia de conclusión, primaria, grupo de edades 12-13 (% del total)')
plt.ylabel('Porcentaje')
plt.xlabel('Años')

plt.show()
```

```python
df_top5.index = df_top5.index.map(int) # let's change the index values of df_top5 to type integer for plotting
df_top5.plot(kind='line', 
             stacked=False,
             figsize=(20, 5), # pass a tuple (x, y) size
             )

plt.title('Tendencia de conclusión, primaria, grupo de edades 12-13 (% del total)')
plt.ylabel('Porcentaje')
plt.xlabel('Años')

plt.show()
```
